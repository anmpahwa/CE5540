# Lecture 10: Hypothesis Testing

```{note}
This lecture introduces the rules of decision-making under uncertainty - Hypothesis Testing.
```

---

This lecture introduces hypothesis testing—a framework for making decisions between competing hypotheses under uncertainty. These hypotheses are typically:

- The null hypothesis ($H_o$): the default or status quo assumption, and

- The alternative hypothesis ($H_1$): the competing claim we seek to test.

A key concept in hypothesis testing is the power of a test (denoted by \beta), which refers to the probability of correctly rejecting a false null hypothesis. In contrast, failing to reject a false null hypothesis results in a Type II error, which occurs with probability 1 − \beta. Further, the test might reject a true null hypothesis, resulting in a Type I error. The probability of this error is denoted by \alpha and is also called the significance level of the test.

An ideal test is therefore one that strikes a balance: it keeps the significance level (\alpha) low to minimize Type I errors, while maintaining high power (\beta) to reduce Type II errors. However, in practice, there's often a trade-off between the two, and thus hypothesis tests must have contextual signficance.

- Airport Security: Consider a metal detector employed at an airport to detect the presence of weapons. In this context, the null hypothesis refers to the assumption that the passenger is not carrying a weapon, while the alternative hypothesis is that the passenger is carrying a weapon. A Type I error occurs when an innocent passenger is incorrectly flagged, causing unnecessary alarm and delay. A Type II error occurs when a dangerous passenger is not detected, posing a significant security threat. Therefore, in this scenario, it is generally preferable to tolerate more Type I errors than to risk a high Type II error, since the consequences of failing to detect a weapon are far more severe.

- Justic System: In a legal trial, the null hypothesis assumes that the defendant is innocent, while the alternative hypothesis asserts that the defendant is guilty. A Type I error corresponds to convicting an innocent person, while a Type II error means acquitting someone who is guilty. Most legal systems prioritize minimizing Type I errors, reflecting the foundational principle that it is better to let a guilty person go free than to wrongly convict an innocent one.

---

**Test Yourself**

Context: According to regulatory standards issued by the Ministry of Road Transport and Highways (MoRTH), all new passenger vehicles sold on or after July 1, 2019, must be equipped with a speed warning system. This system emits intermittent beeps when the vehicle exceeds 80 km/h, and a continuous beep when the speed crosses 120 km/h. Formulate a hypothesis test to evaluate the effectiveness of this policy using high-speed crash data for vehicles with and without such a warning system.

1. The null hypothesis of the test is?

<form id="Q1">
  <label><input type="radio" name="q1" value="a" onchange="checkRadioQ1()"> The warning system has no effect on high-speed vehicle crash </label><br>
  <label><input type="radio" name="q1" value="b" onchange="checkRadioQ1()"> The warning system has lead to a reduction in high-speed vehicle crash </label><br>
</form>
<p id="feedbackQ1"></p>
<script>
function checkRadioQ1() {
  const options = document.getElementsByName("q1");
  let selectedValue = "";
  for (const option of options) {
    if (option.checked) {
      selectedValue = option.value;
      break;
    }
  }
  const feedback = document.getElementById("feedbackQ1");
  if (selectedValue === "a") {
    feedback.innerHTML = "✅ Correct! Consequently, the alternate hypothesis is that the warning system has lead to a reduction in high-speed vehicle crash.";
  } else {
    feedback.innerHTML = "❌ Incorrect. In general, the null hypothesis is chosen to represent the default, status quo, or simpler assumption—typically one that assumes no effect, no difference, or no relationship. This makes it easier to test whether sufficient evidence exists to reject it in favor of a more complex or specific alternative hypothesis.";
  }
}
</script>

2. A Type I error occurs if

<form id="Q2">
  <label><input type="radio" name="q2" value="a" onchange="checkRadioQ1()"> The test concludes that the speed warning system reduces high-speed crashes when in reality it has no effect </label><br>
  <label><input type="radio" name="q2" value="b" onchange="checkRadioQ1()"> The test concludes that the speed warning system does not reduce high-speed crashes, when in fact it does </label><br>
</form>
<p id="feedbackQ2"></p>
<script>
function checkRadioQ2() {
  const options = document.getElementsByName("q2");
  let selectedValue = "";
  for (const option of options) {
    if (option.checked) {
      selectedValue = option.value;
      break;
    }
  }
  const feedback = document.getElementById("feedbackQ2");
  if (selectedValue === "a") {
    feedback.innerHTML = "✅ Correct! In other words, the test wrongly attributes safety improvements to the warning system, possibly justifying further investment or regulatory expansion into speed-based warning systems without actual safety benefits.";
  } else {
    feedback.innerHTML = "❌ Incorrect. Type I errors refer to rejection of a true null hypothesis.";
  }
}
</script>

3. A Type II error occurs if

<form id="Q3">
  <label><input type="radio" name="q3" value="a" onchange="checkRadioQ1()"> The test concludes that the speed warning system reduces high-speed crashes when in reality it has no effect </label><br>
  <label><input type="radio" name="q3" value="b" onchange="checkRadioQ1()"> The test concludes that the speed warning system does not reduce high-speed crashes, when in fact it does </label><br>
</form>
<p id="feedbackQ3"></p>
<script>
function checkRadioQ3() {
  const options = document.getElementsByName("q3");
  let selectedValue = "";
  for (const option of options) {
    if (option.checked) {
      selectedValue = option.value;
      break;
    }
  }
  const feedback = document.getElementById("feedbackQ3");
  if (selectedValue === "b") {
    feedback.innerHTML = "✅ Correct! In other words, the test overlooks a genuinely effective safety intervention, potentially leading to underinvestment in the policy or its premature dismissal despite its positive impact on road safety..";
  } else {
    feedback.innerHTML = "❌ Incorrect. Type II errors refer to failure of rejection of a false null hypothesis.";
  }
}
</script>
